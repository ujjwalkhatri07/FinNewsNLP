- The dataframe is not in correct shape. So we have to reshape the dataframe.
- There are still dictionaries in the column.

```python
# Reshaping the dataframe
rows = []
for ticker, categories in df.items():
    for category, entries in categories.items():
        for entry in entries:
            row = {'Ticker': ticker, 'Category': category, 'Item': entry['item'], 'Title': entry['title']}
            rows.append(row)

ndf= pd.DataFrame(rows)
```

# Text preprocesing

```python
# importing the necessary libraries for text preprocessing
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
ndf
```

```python
# Text preprocessing
stop_words = set(stopwords.words('english'))
ps = PorterStemmer()
def preprocess_text(text):
    text = re.sub('[^a-zA-Z]', ' ', text)  # Removing non-alphabetic characters
    words = text.lower().split()  # Converting to lowercase and splitting into words
    words = [ps.stem(word) for word in words if word not in stop_words]  # Removing stop words and applying stemming
    return ' '.join(words)

ndf['Title'] = ndf['Title'].apply(preprocess_text)
```

# Word Vectorization Methods

## CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Count vectorization
count_vectorizer = CountVectorizer()
X_count = count_vectorizer.fit_transform(ndf['Title'])

# Display word cloud for Count Vectorization
wordcloud_count = WordCloud(width=800, height=400, background_color='white').generate(' '.join(count_vectorizer.get_feature_names_out()))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_count, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Count Vectorization')
plt.show()
```

- It represents the frequency or importance of the text on the basis of size of each word. 
- From the above word cloud from count vectorization we can see that there are most used words like aapl, new, abbvi and abev. 

## TFIDFvectorizer

```python
# Tfidf Vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer


tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(ndf['Title'])

# Display word cloud for TF-IDF Vectorization
wordcloud_tfidf = WordCloud(width=800, height=400, background_color='white').generate(' '.join(tfidf_vectorizer.get_feature_names_out()))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_tfidf, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for TF-IDF Vectorization')
plt.show()
```

- TF-IDF emphasizes terms that are important in a particular document but not necessarily common across the entire dataset.
- From the above word cloud from TF-IDF we can see that there are more number of most important words than in count vectoriation. The most important words in the above word colud are abbvi, accord, acceler, aapl, account, new, access and abev.

- Comparing the two word clouds, the count vectoriation counts the frequency of each word in the document and represents each document as vector of word counts. It treats all words equally in terms of importance and it does not scale the counts based on the document length.
- Whereas Tf-Idf takes into account not ony the term frequency but it also takes the importance of term in the entire corpus and represents each document aas a vector of TF-IDF values. It scales the term frequency based on the inverse document frequency.

# Model Building
## NaiveBayes
- Training set: 80%
- Testing set: 20%

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
# Train-test split for Count Vectorization
X_train_count, X_test_count, y_train, y_test = train_test_split(X_count, ndf['Category'], test_size=0.2, random_state=42)

# Train-test split for TF-IDF Vectorization
X_train_tfidf, X_test_tfidf, _, _ = train_test_split(X_tfidf, ndf['Category'], test_size=0.2, random_state=42)

# Model 1: Multinomial Naive Bayes with Count Vectorization
nb_model_count = MultinomialNB()
nb_model_count.fit(X_train_count, y_train)
nb_train_predictions_count = nb_model_count.predict(X_train_count)
nb_test_predictions_count = nb_model_count.predict(X_test_count)

# Model 2: Multinomial Naive Bayes with TF-IDF Vectorization
nb_model_tfidf = MultinomialNB()
nb_model_tfidf.fit(X_train_tfidf, y_train)
nb_train_predictions_tfidf = nb_model_tfidf.predict(X_train_tfidf)
nb_test_predictions_tfidf = nb_model_tfidf.predict(X_test_tfidf)

# Evaluate models
print("Naive Bayes Train Accuracy with Count Vectorization:", accuracy_score(y_train, nb_train_predictions_count))
print("Naive Bayes Test Accuracy with Count Vectorization:", accuracy_score(y_test, nb_test_predictions_count))

print("Naive Bayes Train Accuracy with TF-IDF Vectorization:", accuracy_score(y_train, nb_train_predictions_tfidf))
print("Naive Bayes Test Accuracy with TF-IDF Vectorization:", accuracy_score(y_test, nb_test_predictions_tfidf))

# Classification report for Naive Bayes with Count Vectorization
print("Naive Bayes Classification Report with Count Vectorization:\n", classification_report(y_test, nb_test_predictions_count))

# Classification report for Naive Bayes with TF-IDF Vectorization
print("Naive Bayes Classification Report with TF-IDF Vectorization:\n", classification_report(y_test, nb_test_predictions_tfidf))
```

- The output shows the training and testing accuracy, as well as the classification reports for two models: Naive Bayes with Count Vectorization and Naive Bayes with TF-IDF Vectorization
- Naive Bayes with Count Vectorization:
 - Train Accuracy: 83.19%
 - Test Accuracy: 81.34%
- Naive Bayes with TF-IDF Vectorization:
 - Train Accuracy: 85.02%
 - Test Accuracy: 83.19%
- The model with TF-IDF Vectorization performs slightly better on both the training and testing set which shows that that the additional TF-IDF feature extraction technique is providing some improvement.
- Both the model seems to perform reasonably well, but the TF-IDF Vectorization model shows a slightly better overall accuracy on the test set.

```python
from sklearn.model_selection import learning_curve


# Model 1: Logistic Regression with Count Vectorization
logreg_model_count = LogisticRegression()
logreg_model_count.fit(X_train_count, y_train)

# Model 2: Multinomial Naive Bayes with TF-IDF Vectorization
nb_model_tfidf = MultinomialNB()
nb_model_tfidf.fit(X_train_tfidf, y_train)

# Calculate training and testing accuracy over epochs
train_sizes, train_scores_logreg_count, test_scores_logreg_count = learning_curve(
    logreg_model_count, X_train_count, y_train, cv=5, scoring='accuracy')
train_sizes, train_scores_nb_tfidf, test_scores_nb_tfidf = learning_curve(
    nb_model_tfidf, X_train_tfidf, y_train, cv=5, scoring='accuracy')

# Calculate mean accuracy values over cross-validation folds
mean_train_logreg_count = np.mean(train_scores_logreg_count, axis=1)
mean_test_logreg_count = np.mean(test_scores_logreg_count, axis=1)
mean_train_nb_tfidf = np.mean(train_scores_nb_tfidf, axis=1)
mean_test_nb_tfidf = np.mean(test_scores_nb_tfidf, axis=1)

# Plotting training and testing accuracy curves
plt.figure(figsize=(10, 5))

plt.plot(train_sizes, mean_train_logreg_count, 'bo-', label='Logistic (Count) Train')
plt.plot(train_sizes, mean_test_logreg_count, 'r+-', label='Logistic (Count) Test')

plt.plot(train_sizes, mean_train_nb_tfidf, 'gs-', label='Naive Bayes (TF-IDF) Train')
plt.plot(train_sizes, mean_test_nb_tfidf, 'y^-', label='Naive Bayes (TF-IDF) Test')

plt.title('Training and Testing Accuracy Curves')
plt.xlabel('Training Examples')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
